## Visual EDA

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("../1-Data/Raw/telco.csv")
df.head()
```

```python
df.columns = df.columns.str.strip().str.replace(" ", "_")
```

```python
df['Churn_Label'] = df['Churn_Label'].map({'Yes': 1, 'No': 0})
```

## Churn Distribution

```python
# Count plot
sns.countplot(x='Churn_Label', data=df)
plt.title("Customer Churn Distribution")
plt.show()
```

```python
#Contract Type vs Churn
sns.countplot(x='Contract', hue='Churn_Label', data=df)
plt.title("Churn by Contract Type")
plt.xticks(rotation=45)
plt.show()
```

```python
#Total revenue vs Churn
sns.boxplot(x='Churn_Label', y='Total_Revenue', data=df)
plt.title("Revenue Distribution by Churn Status")
plt.show()
```

```python
sns.boxplot(x='Churn_Label', y='Tenure_in_Months', data=df)
plt.title("Tenure Distribution by Churn Status")
plt.show()
```

```python
sns.countplot(x='Payment_Method', hue='Churn_Label', data=df)
plt.title("Churn by Payment Method")
plt.xticks(rotation=45)
plt.show()
```
