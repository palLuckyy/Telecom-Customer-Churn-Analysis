### Customer Churn Analysis & Business Intelligence Dashboard

```python
import pandas as pd

df = pd.read_csv("../1-Data/Raw/telco.csv")

df.head()
```

### Chek data is clean or not

```python
df.shape
df.info()
df.isnull().sum()
df.duplicated().sum()
```

### Proper Cleaning

```python
# 1️⃣ Standardize column names (very important)
df.columns = df.columns.str.strip().str.replace(" ", "_")

# 2️⃣ Handle Offer column
df['Offer'] = df['Offer'].fillna("No_Offer")

# 3️⃣ Handle Internet_Type
df['Internet_Type'] = df['Internet_Type'].fillna("No_Internet")

# 4️⃣ Handle churn related columns
df['Churn_Category'] = df['Churn_Category'].fillna("No_Churn")
df['Churn_Reason'] = df['Churn_Reason'].fillna("Not_Applicable")

# 5️⃣ Convert categorical Yes/No columns to consistent format
yes_no_cols = df.select_dtypes(include='object').columns

for col in yes_no_cols:
    df[col] = df[col].str.strip()

# 6️⃣ Check again
df.isnull().sum()
```

### Select target columns

```python
target = 'Churn_Label'
```

```python
df['Churn_Label'].value_counts()
```

### Remove useless columns

```python
drop_cols = [
    'Customer_ID',
    'Zip_Code',
    'Latitude',
    'Longitude'
]

df.drop(columns=drop_cols, inplace=True)
```

```python
df.info()
df.isnull().sum()
df.duplicated().sum()
```

```python
df['Churn_Label'] = df['Churn_Label'].map({'Yes': 1, 'No': 0})
```

```python
df['Churn_Label'].value_counts()
df['Churn_Label'].value_counts(normalize=True)*100
```

```python
churn_rate = df['Churn_Label'].mean() * 100
print(f"Churn Rate: {churn_rate:.2f}%")
```

### Business Objective

#### The telecom company is experiencing customer churn.
#### The objective of this analysis is to identify key churn drivers, measure financial impact, and recommend data-driven customer retention strategies.

```python
total_customers = len(df)
total_churned = df['Churn_Label'].sum()
churn_rate = df['Churn_Label'].mean()*100
total_revenue = df['Total_Revenue'].sum()
revenue_lost = df[df['Churn_Label']==1]['Total_Revenue'].sum()

print(f"Total Customers: {total_customers}")
print(f"Total Churned Customers: {total_churned}")
print(f"Churn Rate: {churn_rate:.2f}%")
print(f"Total Revenue: ${total_revenue:,.2f}")
print(f"Revenue Lost Due to Churn: ${revenue_lost:,.2f}")
```

### Insights

```python
revenue_lost / total_revenue * 100
```
